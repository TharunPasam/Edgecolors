```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(10,10))

samples = np.random.randn(30,2)

ax[0][0].scatter(samples[:,0], samples[:,1],
            color='red',
            label='color="red"')

ax[1][0].scatter(samples[:,0], samples[:,1],
            c='red',
            label='c="red"')

ax[0][1].scatter(samples[:,0], samples[:,1],
            edgecolor='red',
            c='red',
            label='c="red"')

ax[1][1].scatter(samples[:,0], samples[:,1],
            edgecolor='green',
            c='1',
            label='color="1.0", edgecolor="0"')

ax[1][2].scatter(samples[:,0], samples[:,1],
            edgecolor='green',
            c='1',
            label='color="1.0", edgecolor="0"')

ax[2][1].scatter(samples[:,0], samples[:,1],
            edgecolor='green',
            c='1',
            label='color="1.0", edgecolor="0"')

ax[0][2].scatter(samples[:,0], samples[:,1],
            edgecolor='blue',
            c='1',
            label='color="1.0", edgecolor="0"')

ax[2][0].scatter(samples[:,0], samples[:,1],
            edgecolor='blue',
            c='1',
            label='color="1.0", edgecolor="0"')

ax[2][2].scatter(samples[:,0], samples[:,1],
            edgecolor='blue',
            c='1',
            label='color="1.0", edgecolor="0"')

for row in ax:
    for col in row:
        col.legend(loc='upper left')

plt.show()
```

```python
import matplotlib.pyplot as plt

import numpy as np
import matplotlib.pyplot as plt

markers = [

'.', # point
',', # pixel
'o', # circle
'v', # triangle down
'^', # triangle up
'<', # triangle_left
'>', # triangle_right
'8', # octogon
'+', #plus
'H', #hexagon
]

plt.figure(figsize=(10, 8))
samples = range(len(markers))


for i in samples:
    plt.plot([i-2, i, i+2], [i, i, i], label=markers[i], marker=markers[i], markersize=20)


# Annotation

plt.title('Marker styles', fontsize=20)
plt.ylim([-1, len(markers)+1])
plt.legend(loc='upper left')


plt.show()
```

```python
 import numpy as np
import matplotlib.pyplot as plt

linestyles = ['None', '--', '-.', '-', ':']

plt.figure(figsize=(8, 5))
samples = range(len(linestyles))


for i in samples:
    plt.plot([i-2, i, i+2], [i, i, i],
             label='"%s"' %linestyles[i],
             linestyle=linestyles[i],
             )


plt.title('Line styles', fontsize=20)
plt.ylim([-1, len(linestyles)+2])
plt.legend(loc='lower right')


plt.show()
```
